```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
```

# make swiss roll (original)

```python
def make_swiss_roll(n_samples=100, noise=0.0):
    """stolen from sklearn"""
    t = 1.5 * np.pi * (1 + 2* np.random.rand(1, n_samples)) # the original generator
    
    x = t * np.cos(t)
    y = 21 * np.random.rand(1, n_samples)
    z = t * np.sin(t)

    X = np.concatenate((x, y, z))
    X += noise * np.random.randn(3, n_samples)
    X = X.T
    t = np.squeeze(t)

    return X, t
```

```python
def sr_viz(x,t):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot('111', projection='3d')
    ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=t, cmap=plt.cm.Spectral)
```

# make half swiss roll test

```python
def test1(n_samples=100, noise=0.0, k=2):
    '''testing the periodic segementation of the swiss roll'''
    t = k * np.pi * np.random.rand(1, n_samples)    
    
    x = t * np.cos(t)
    y = 21 * np.random.rand(1, n_samples)
    z = t * np.sin(t)

    X = np.concatenate((x, y, z))
    X += noise * np.random.randn(3, n_samples)
    X = X.T
    t = np.squeeze(t)

    return X, t
```

```python
def test2(n_samples=100, noise=0.0, k=2):
    '''testing the periodic segementation of the swiss roll'''
    t = np.pi * (k + np.random.rand(1, n_samples))
    
    x = t * np.cos(t)
    y = 21 * np.random.rand(1, n_samples)
    z = t * np.sin(t)

    X = np.concatenate((x, y, z))
    X += noise * np.random.randn(3, n_samples)
    X = X.T
    t = np.squeeze(t)

    return X, t
```

```python
def test3(n_samples=100, noise=0.0, k=2):
    '''constructing 1d partial swiss roll'''
    t = np.pi * (k + np.random.rand(1, n_samples))    
    x = t * np.cos(t)    
    z = t * np.sin(t)

    X = np.concatenate((x, z))
    X += noise * np.random.randn(2, n_samples)
    X = X.T
    t = np.squeeze(t)

    return X, t
```

```python
def test4(n_samples=100, noise=0.0):
    """stolen from sklearn"""
    t = 1.5 * np.pi * (2 + (2/3) * np.random.rand(1, n_samples)) # the original generator
    
    x = t * np.cos(t)
    y = 21 * np.random.rand(1, n_samples)
    z = t * np.sin(t)

    X = np.concatenate((x, y, z))
    X += noise * np.random.randn(3, n_samples)
    X = X.T
    t = np.squeeze(t)

    return X, t
```

# visulization

```python
x, t = test3(n_samples=6000)
plt.scatter(x[:,0], x[:,1], c=t, cmap=plt.cm.Spectral)
plt.title("1D partial SR: [2pi,3pi]")
```

```python
x, t = make_swiss_roll(n_samples=6000)
sr_viz(x,t)
plt.title("The original version of SR from sklearn")
```

```python
x, t = test4(n_samples=6000)
sr_viz(x,t)
```

```python
x, t = test2(n_samples=6000, k=2)
sr_viz(x,t)
plt.title('[2pi,3pi]')
```

```python
x, t = test1(n_samples=6000, k=1)
sr_viz(x,t)
plt.title('[0,pi]')
```

```python
x, t = test1(n_samples=6000, k=2)
sr_viz(x,t)
plt.title('[0,2pi]')
```
